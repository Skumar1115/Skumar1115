```python
import re
def process_text(Notes):
    return " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])", " ",Notes.lower()).split())
```

```python
data['text']=data['Notes'].apply(process_text)
```

```python

from tensorflow.keras.preprocessing.text import one_hot
voc_size=5000
onehot_repre=[one_hot(words,voc_size)for words in data['text']]
```

```python
onehot_repre
```

```python

sent_length=20
embedding_docs=pad_sequences(onehot_repre,padding='pre',maxlen=sent_length)
```

```python
x=np.array(embedding_docs)
x
```

```python
y=data['s1_etr_time']
y
```

```python
sent_length=20
voc_size=5000
embedding_vector_features=40
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.2))
model1.add(Dense(4,activation='relu'))
model1.add(Dense(11,activation='softmax'))
model1.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.22,random_state=120,)
```
